```python
import numpy as np
import pandas as pd
import scipy

import sklearn
import sklearn.datasets

%pylab inline

import warnings
warnings.filterwarnings('ignore')
```

```python
samples1 = scipy.random.multivariate_normal(mean=[5,0], cov=[[1,2], [2,1]], size=(2000,)) 
labels1 = np.ones(samples1.shape[0])
samples2 = scipy.random.multivariate_normal(mean=[-5,1], cov=[[1,0], [0,1]], size=(2000,))
labels2 = -np.ones(samples2.shape[0])
```

```python
samples = np.vstack([samples1, samples2])
labels = np.hstack([labels1, labels2])
```
